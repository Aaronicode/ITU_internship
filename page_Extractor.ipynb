```python
import PyPDF2

def extract_and_create_pdf(input_pdf_path, output_pdf_path, page_number):
    # Open the PDF file
    with open(input_pdf_path, 'rb') as file:
        # Create a PDF reader object
        pdf_reader = PyPDF2.PdfReader(file)


        if 0 < page_number <= len(pdf_reader.pages):
            # Create a new PDF writer object
            pdf_writer = PyPDF2.PdfWriter()

            # Get the specified page
            page = pdf_reader.pages[page_number - 1]

            # Add the page to the new PDF
            pdf_writer.add_page(page)

            # Write the new PDF to the output file
            with open(output_pdf_path, 'wb') as output_file:
                pdf_writer.write(output_file)

            print(f"Page {page_number} extracted and saved to {output_pdf_path}")
        else:
            print(f"Invalid page number: {page_number}. Please choose a valid page.")

# Example usage
input_pdf_path = '/content/drive/MyDrive/T-REC-Y.Sup71-202207-I!!PDF-E.pdf'
output_pdf_path = 'test_doc_9.pdf'
page_number_to_extract = 50
for i in range(10):
  page_number_to_extract+=1
  output_pdf_path=f"test_doc{page_number_to_extract}"
  extract_and_create_pdf(input_pdf_path, output_pdf_path, page_number_to_extract)
```
